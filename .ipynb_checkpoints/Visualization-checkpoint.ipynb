```python
df = pd.read_csv("data3.csv")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X = df['Marks'].values
Y = df["NextMarks"].values
Z = df['Applied'].values
#ax.plot_surface(X, Y, Z,linewidth=0, antialiased=False)
ax.scatter(X, Y, Z, c=Z, marker='o',vmin=0,vmax=1)
#ax.plot_surface(X, Y, Z)
#ax.plot_trisurfr(X,Y,Z)
plt.show()
fig.show()
```
